```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_img = cv2.imread("./goruntu.jpg")
```

```python
print(f"Görüntünün yüksekliği = {my_img.shape[0]}")
print(f"Görüntünün genişliği = {my_img.shape[1]}")
print(f"Görüntünün kanal sayısı = {my_img.shape[2]}")
```

```python
#Bir görüntünün yüksekliği satır sayısıyla, genişliği sütun sayısıyla doğru orantılıdır.
#Bir görüntü (normal şartlarda) 0 ile 255 arasında 256 adet tamsayıyla ifade edilir.
#0, siyah rengine; 255 ise beyaz rengine karşılık gelir.
#RGB kanalında siyah rengi: [0,0,0] ile ifade edilir.
#RGB kanalında beyaz rengi: [255,255,255] ile ifade edilir.
#Gri rengi, siyah ve beyazın eşit oranda karıştırılmasıyla oluşur.
#Bu durumda 0 ve 255'i toplayıp 2'ye bölersek, sonuç bize gri renginin sayısal karşılığını verir.
#BT.601 standardı: İnsan gözü en iyi yeşil rengini sonra kırmızıyı en sonda maviyi ayırt eder.
#BT.601 standardı, üç kanallı RGB görüntüyü tek kanallı gri seviyeye çekmek için kullanılan ağırlıklandırma formülüdür.
#Bu formülde en büyük ağırlık yeşil rengine, sonra kırmızıya sonra da maviye verilir.
#Her bir piksel kendi ağırlığıyla çarpılır ve sonunda hepsi toplanarak 3 kanaldan tek kanallı bir sonuç elde edilir.
```

```python
print(f"Matristeki en düşük piksel değeri = {np.min(my_img)}")
print(f"Matristeki en yüksek piksel değeri = {np.max(my_img)}")
```

```python
my_img[600,500] #Burada görüntünün 600. satır, 500. sütunda hangi BGR piksel değerleri var onu gösterir
```

```python
yeni_2 = cv2.cvtColor(my_img,cv2.COLOR_BGR2RGB) #Burada BGR'dan RGB'ye renk dönüşümü yapılmış. Oluşan RGB görüntüyü yeni_2 isimli bir değişkene atanmıştır.
```

```python
cv2.imwrite("deneme.jpeg",yeni_2) #yeni_2 görüntüsünü cv2.imwrite fonksiyonu ile "deneme.jpeg" olarak diske kaydedilmiştir.
```

```python
gri_img = cv2.cvtColor(my_img,cv2.COLOR_BGR2GRAY)
```

```python
print(f"orijinal görüntünün boyutları ve kanal sayısı = {my_img.shape}")
print(f"gri görüntünün boyutları ve kanal sayısı = {gri_img.shape}")
```

```python
#İlk olarak BGR kanal sıralamasında okuduğumuz renkli görüntünün, sırasıyla; RGB renk kanalı sıralamasıyla dönüştürülmesi (convert edilmesi), ve sonra gri seviye olarak dönüştürülmesi.
my_rgb_img = cv2.cvtColor(my_img,cv2.COLOR_BGR2RGB)
my_gray_img = cv2.cvtColor(my_rgb_img, cv2.COLOR_RGB2GRAY)
```

```python
#plt.subplot(satır_sayısı, sütun_sayısı; kaçıncı_görüntü_olduğu)
#Örneğin: plt.subplot(3,2,4) kodunda satır sayısı ve sütun sayısını çarparsak bize toplam gösterilecek görüntü sayısını verir.
#Toplam görüntü yukarıdaki örnek için; 3*2 = 6 görüntü gösterilecek. Sondaki 4 sayısı ise o anki plot'ta hangi görüntünün gösterildiğini ifade eder.
#Yani 6 görüntü arasından 4. görüntüyü gösterdiğini açıklıyor.

fig = plt.figure(figsize=(10,7)) #görüntünün boyutu

plt.subplot(1, 2, 1)
plt.imshow(my_rgb_img)
#plt.axis('off') #koordinat çizgilerini kapatır.
plt.title("RGB Image")

plt.subplot(1, 2, 2)
plt.imshow(my_gray_img,cmap="gray") 
#plt.axis('off')
plt.title("Gray Image")

#"cmap" parametresi, fonksiyonun tek kanallı bir görüntüyü alıp; en düşük değeri siyaha, en yüksek değeri beyaza eşlemesini sağlar.
```
